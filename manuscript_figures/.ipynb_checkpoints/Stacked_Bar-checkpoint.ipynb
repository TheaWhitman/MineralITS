```python
%%R
library(phyloseq)
library(plyr)
library(dplyr); library(tidyr)
library(ggplot2)
library(rwantshue)
```

```python
%%R
physeq = import_biom("../../SeqData/otu_table.tax.meta.biom", "../../SeqData/trees/fulltree.tre", parseFunction = parse_taxonomy_greengenes)
```

```python
%%R
physeq.noSmall <- subset_samples(physeq,sample_sums(physeq)>=2000)
```

```python
%%R
mdf = psmelt(physeq.noSmall)
```

```python
%%R -w 900
d.plot = mdf %>%
    group_by(Sample) %>%
    mutate(Abundance = Abundance / sum(Abundance)) %>%
    group_by(Sample, Phylum, Trtmt, Month) %>%
    summarize(Abundance = sum(Abundance)) %>%
    arrange(Trtmt)

mean.Abund = ddply(d.plot, .(Phylum), summarize, M = mean(Abundance))
# Calculate the phyla's mean abundances
d.plot$Phylum = factor(d.plot$Phylum, levels=mean.Abund[order(-mean.Abund$M),]$Phylum)
# Order the Phyla by abundance

trtmt.order <- c("S","Q","F","K","H")
d.plot$Trtmt = factor(d.plot$Trtmt, levels=trtmt.order)


d.plot<-d.plot[with(d.plot, order(Month, Trtmt, Sample, Phylum)),]
# Sort the whole dataframe by this phylum order

scheme <- iwanthue()
colors = scheme$hex(length(unique(d.plot$Phylum)), color_space = hcl_presets$pastel)

p = ggplot(d.plot, aes(x = Sample, y = Abundance, fill = Phylum))

p = p + facet_grid(. ~ Trtmt, scales = "free_x", space = "free")

p = p + scale_fill_manual(values = colors)

p = p + geom_bar(stat = "identity", width = 0.70)

p = p + guides(fill = guide_legend(ncol = 2))

p = p + theme_bw()

p = p + theme(axis.text.x = element_blank())

p
```

```python
#%%R

N=500
Phyla = levels(mdf$Phylum)

NPhylum <- function (x){
    m = c()
    for (i in x){
       L <- length(mdf$Phylum[mdf$Phylum==i])
    if (L > N) m = c(m,i)
        }
        m
    }

Include<-as.factor(NPhylum(Phyla))
mdf.trim<-mdf[mdf$Phylum %in% Include,]
mdf.trim$Phylum<-factor(mdf.trim$Phylum)
mdf.trim$Phylum <- droplevels(mdf.trim$Phylum)
levels(mdf.trim$Phylum)
```
